## 1. Introduction

In March 2024, devastating floods occurred in East Africa, particularly in Kenya and Tanzania, with destructive effects on infrastructure, property, and human lives. According to news reports:
- **Kenya**: ~40 deaths and ~40,000 displaced.
- **Tanzania**: ~155 deaths.
- **Burundi**: ~100,000 displaced.

The floods resulted from several days of heavy downpours that overwhelmed the region's rivers. The lack of adequate ground observation stations makes it challenging to analyze such events in this region. Satellite observations, such as the H SAF precipitation products, offer a critical tool for studying these phenomena and their impacts.

This notebook utilizes the **H64 H SAF satellite-derived precipitation products** to study the spatial and temporal patterns of the March-April 2024 heavy rainfall in East and Central Africa.

---

## 2. Objectives

The primary goals of this study are:
1. To demonstrate the utility of the H SAF H64 precipitation products in analyzing heavy rain events.
2. To estimate the spatial and temporal extent of rainfall during March-April 2024 using H60 and H64 H SAF products.
3. To assess the flood’s impact on land cover and identify vulnerable areas for future flood mitigation efforts.

---

## 3. Study Area

The study focuses on countries in **East and Central Africa**, including:
- **East Africa**: Zambia, Burundi, Kenya, Malawi, Rwanda, Uganda, Tanzania, and Mozambique.
- **Central Africa**: Democratic Republic of Congo (DRC).
The boundaries of these regions are defined using a shapefile and visualized alongside the rainfall data.

### Key Geographical Features:
- **Lakes**: Proximity to Lake Victoria and Lake Tanganyika plays a crucial role in the hydrological system but also exacerbates flood risks.
- **Seasonal Rainfall**: Long rains occur from **March to May**, significantly contributing to flooding.
- **Flood Vulnerability**: Landlocked countries like Rwanda and Burundi are particularly vulnerable due to their geography and climate.

## 4. Materials and Methods

### 4.1 Data Sources
- **H SAF Products**:
  - **H64**: Daily precipitation on a regular grid (0.25° x 0.25°).
  - **H60**: Instantaneous precipitation product at 15 minutes interval.

- **Satellite and Auxiliary Data**:
  - Sentinel-1 Ground Range Detected (GRD) data for flood extent mapping.
  - ESA WorldCover 10m v100 for land classification.


### 4.2 Tools
- Google Earth Engine for data integration and visualization.

### 4.3 Workflow
1. **Initial Analysis**:
   - H60 was used to determine precipitation movement and to identify regions of heavy rainfall.
2. **Rainfall Mapping**:
   - H64 product was were employed to estimate the spatial and temporal extent of rainfall using the previous notebooks. Figure 1 shows a plot of the daily precipitations.
3. **Flood Extent Mapping**:
   - Sentinel-1 GRD data was used to estimate flooded areas.
4. **Impact Assessment**:
   - Land cover types were analyzed to assess flood vulnerability.

## 5. Spatial and Temporal Patterns of Precipitation

### 5.1 Mean Rainfall Analysis
Analysis of the H64 precipitation product revealed that the overall mean rainfall for the study area during the defined period was **6.90 mm**, with a median rainfall of **2.03 mm**. The data displayed significant variability, with a standard deviation of **15.95 mm**, indicating extreme events such as the heavy downpours leading to flooding. The histogram indicates that rainfall is highly skewed, with most values clustering around low intensities, while a few extreme values dominate the dataset. The skewness of **6.24** highlights the dominance of high-intensity rainfall events in March-April 2024.

**Rainfall Statistics**:
- Overall Mean Rainfall: **6.90 mm**
- Median Rainfall: **2.03 mm**
- Standard Deviation: **15.95 mm**
- Skewness: **6.24**

![Log_frequency_distribution](sample/H64_log_freq.png)
*Figure 2: Log frequency distribution of the March-April 2024 precipitation in the AOI*



### 5.2 Spatial Distribution
A spatial map of mean rainfall reveals regions within the Area of Interest (AOI) with significantly high accumulated rainfall, particularly along permanent surface water bodies. Notably, areas surrounding **Lake Victoria** and portions of **Kenya** and **Tanzania** experienced the heaviest rainfall during March-April 2024. These findings align with reports of severe flooding in the region.  

![Mean Rainfall Map](sample/H64_Mean_Rainfall.png)

### 5.3 Temporal Trends
The time series plot demonstrates that extreme rainfall events occurred within a short window, coinciding with documented flood events in the region. Peaks in rainfall intensity during March and April indicate the progression of the long rains, critical for understanding flood triggers.
